```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('2015.csv')
```

```python
data = data[
    [
        'DIABETE3',
        '_RFHYPE5',
        'TOLDHI2',
        '_CHOLCHK',
        '_BMI5',
        'SMOKE100',
        'CVDSTRK3',
        '_MICHD',
        '_TOTINDA',
        '_FRTLT1',
        '_VEGLT1',
        '_RFDRHV5',
        'HLTHPLN1',
        'MEDCOST',
        'GENHLTH',
        'MENTHLTH',
        'PHYSHLTH',
        'DIFFWALK',
        'SEX',
        '_AGEG5YR',
        'EDUCA',
        'INCOME2',
        '_RACE_G1',
        '_STATE'
    ]
]
```

```python
data.shape
```

### Data Preprocessing

```python
data['_RFHYPE5'] = data['_RFHYPE5'].replace({1: 0, 2: 1})
data = data[data._RFHYPE5 != 9]

data['TOLDHI2'] = data['TOLDHI2'].replace({2: 0})
data = data[data.TOLDHI2 != 7]
data = data[data.TOLDHI2 != 9]

data['_CHOLCHK'] = data['_CHOLCHK'].replace({3: 0, 2: 0})
data = data[data._CHOLCHK != 9]

data['_BMI5'] = data['_BMI5'].div(100).round(0)

data['SMOKE100'] = data['SMOKE100'].replace({2: 0})
data = data[data.SMOKE100 != 7]
data = data[data.SMOKE100 != 9]

data['CVDSTRK3'] = data['CVDSTRK3'].replace({2: 0})
data = data[data.CVDSTRK3 != 7]
data = data[data.CVDSTRK3 != 9]

data['_MICHD'] = data['_MICHD'].replace({2: 0})

data['_TOTINDA'] = data['_TOTINDA'].replace({2: 0})
data = data[data._TOTINDA != 9]

data['_FRTLT1'] = data['_FRTLT1'].replace({2: 0})
data = data[data._FRTLT1 != 9]

data['_VEGLT1'] = data['_VEGLT1'].replace({2: 0})
data = data[data._VEGLT1 != 9]

data['_RFDRHV5'] = data['_RFDRHV5'].replace({1: 0, 2: 1})
data = data[data._RFDRHV5 != 9]


data['HLTHPLN1'] = data['HLTHPLN1'].replace({2: 0})
data = data[data.HLTHPLN1 != 7]
data = data[data.HLTHPLN1 != 9]

data['MEDCOST'] = data['MEDCOST'].replace({2: 0})
data = data[data.MEDCOST != 7]
data = data[data.MEDCOST != 9]

data = data[data.GENHLTH != 7]
data = data[data.GENHLTH != 9]

data['MENTHLTH'] = data['MENTHLTH'].replace({88: 0})
data = data[data.MENTHLTH != 77]
data = data[data.MENTHLTH != 99]

data['PHYSHLTH'] = data['PHYSHLTH'].replace({88: 0})
data = data[data.PHYSHLTH != 77]
data = data[data.PHYSHLTH != 99]

data['DIFFWALK'] = data['DIFFWALK'].replace({2: 0})
data = data[data.DIFFWALK != 7]
data = data[data.DIFFWALK != 9]

# Original: 1 - Male  2 - Female
data['SEX'] = data['SEX'].replace({2: 0})

data = data[data._AGEG5YR != 14]

data = data[data.EDUCA != 9]

data = data[data.INCOME2 != 77]
data = data[data.INCOME2 != 99]

data = data[data._RACE_G1 != 9]

data = data[data._STATE != 99]
```

### Binarizing Target Variable

```python
data['DIABETE3'].value_counts()
```

```python
data = data[data.DIABETE3 != 7]
data = data[data.DIABETE3 != 9]

data['DIABETE3'] = data['DIABETE3'].replace({2: 0, 3: 0, 4: 1})
```

```python
data['DIABETE3'].value_counts()
```

### Cleaning null values by removing rows with any null value

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

### Renaming

```python
data = data.rename(columns={
    'DIABETE3': 'Diabetes',
    '_RFHYPE5': 'PressAlta',
    'TOLDHI2': 'ColestAlto',
    '_CHOLCHK': 'VerifColest',
    '_BMI5': 'IMC',
    'SMOKE100': 'Fumante',
    'CVDSTRK3': 'Derrame',
    '_MICHD': 'DoencaCardiaca',
    '_TOTINDA': 'AtivFisica',
    '_FRTLT1': 'Frutas',
    '_VEGLT1': 'Vegetais',
    '_RFDRHV5': 'ConsumoAlcool',
    'HLTHPLN1': 'AssistMedica',
    'MEDCOST': 'SemMedicoCusto',
    'GENHLTH': 'SaudeGeral',
    'MENTHLTH': 'SaudeMental',
    'PHYSHLTH': 'SaudeFisica',
    'DIFFWALK': 'DificCaminhar',
    'SEX': 'Sexo',
    '_AGEG5YR': 'Idade',
    'EDUCA': 'Escolaridade',
    'INCOME2': 'Renda',
    '_RACE_G1': 'Etnia',
    '_STATE': 'Estado'
})
```
